```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn import metrics
```

```python
data = pd.read_csv("E:/Academics/data/wisc_bc_data.csv")
```

```python
data.head()
```

```python
del data["id"]
```

```python
data.head()
```

```python
data1 = data.iloc[:, 1:]
```

```python
data1.head()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
data_scaled = min_max_scaler.fit_transform(data1)
```

```python
data2 = pd.DataFrame(data_scaled)
```

```python
data2.head()
```

```python
data2[30] = data["diagnosis"]
```

```python
data2.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
label = label_encoder.fit_transform(data2[30])
```

```python
label
```

```python
data2[30] = label
```

```python
data2.head()
```

```python
train_x, test_x, train_y, test_y = train_test_split(data2.iloc[:, :-1],data2[30], test_size = 0.2)
```

```python
test_x.head()
```

```python
test_y.head()
```

```python
knn2 = knn(n_neighbors=21)
```

```python
knn2.fit(train_x, train_y)
```

```python
y_pred = knn2.predict(test_x)
```

```python
metrics.confusion_matrix(y_pred, test_y)
```

```python
metrics.accuracy_score(y_pred,test_y)
```

```python
train_x.shape
```

```python
knn_class = []
for i in range(1, 455):
    knn_class = knn(n_neighbors = i)
    knn_class.fit(train_x, train_y)
    y_pred = knn_class.predict(test_x)
    print("Accuracy of k =" , i)
    print(metrics.accuracy_score(y_pred, test_y))
    
```
